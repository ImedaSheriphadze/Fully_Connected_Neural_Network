Пример математической игры средней сложности на Python под названием «**Математическая викторина**»

В этой игре **Math Quiz** программа предлагает игроку серию математических вопросов. Каждый вопрос состоит из двух случайных чисел от 1 до 20 и случайного арифметического оператора (+, -, *, /). Игроку необходимо ввести правильный ответ на каждый вопрос.

Игра состоит из **5** вопросов. После каждого ответа программа предоставляет обратную связь, указывающую, является ли ответ правильным или неправильным. В конце игры программа выводит счет игрока из общего количества вопросов.

Для запуска игры запустите программу. Программа предложит вам математические вопросы, и вам нужно ввести ответы. После ответа на все вопросы программа отобразит ваш балл.

Не стесняйтесь настраивать количество вопросов, диапазон чисел или добавлять более сложные математические операции, чтобы сделать игру более сложной.

```python
import random

def play_math_quiz():
    num_questions = 5
    score = 0
    
    print("Добро пожаловать на математическую викторину!")
    print("Ответьте на следующие математические вопросы:")
    print("-----------------------------------")
    
    for _ in range(num_questions):
        num1 = random.randint(1, 20)
        num2 = random.randint(1, 20)
        operator = random.choice(['+', '-', '*', '/'])
        
        question = f"Что {num1} {operator} {num2}? "
        user_answer = float(input(question))
        
        if operator == '+':
            actual_answer = num1 + num2
        elif operator == '-':
            actual_answer = num1 - num2
        elif operator == '*':
            actual_answer = num1 * num2
        elif operator == '/':
            actual_answer = num1 / num2
        
        if user_answer == actual_answer:
            print("Правильный!")
            score += 1
        else:
            print(f"Неправильно! Правильный ответ {actual_answer}.")
        
        print("-----------------------------------")
    
    print(f"Игра закончена! ваш счет {score} снаружи {num_questions}.")

# Запустить игру
play_math_quiz()
```
