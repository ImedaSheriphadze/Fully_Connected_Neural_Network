Вот уникальная программа, которая демонстрирует возможности персептрона по классификации образов, решая задачу **XOR**. Он отображает два класса на точечной диаграмме, вычисляет процент правильности сети, строит соответствующий график и показывает принцип работы задачи **XOR**.

```python
import numpy as np
import matplotlib.pyplot as plt

# Класс персептрона
class Perceptron:
    def __init__(self, input_size):
        self.weights = np.random.rand(input_size + 1)

    def activate(self, x):
        return 1 if x >= 0 else 0

    def predict(self, inputs):
        inputs = np.append(inputs, 1)  # add bias
        summation = np.dot(inputs, self.weights)
        return self.activate(summation)

    def train(self, training_inputs, labels, epochs, learning_rate):
        accuracy = []
        for epoch in range(epochs):
            correct_predictions = 0
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                if prediction == label:
                    correct_predictions += 1
                self.weights += learning_rate * (label - prediction) * np.append(inputs, 1)
            accuracy.append(correct_predictions / len(labels) * 100)

        return accuracy, accuracy[-1]

# Пример XOR-задачи
training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels = np.array([0, 1, 1, 0])

# Создать и обучить персептрон
perceptron = Perceptron(input_size=2)
epochs = 1000
learning_rate = 0.1
accuracy, correct_percentage = perceptron.train(training_inputs, labels, epochs, learning_rate)

# Создание точек данных для двух классов
x_class0 = []
y_class0 = []
x_class1 = []
y_class1 = []

for i in range(len(training_inputs)):
    if labels[i] == 0:
        x_class0.append(training_inputs[i][0])
        y_class0.append(training_inputs[i][1])
    else:
        x_class1.append(training_inputs[i][0])
        y_class1.append(training_inputs[i][1])

# Постройте два класса
plt.scatter(x_class0, y_class0, c='red', label='Класс 0')
plt.scatter(x_class1, y_class1, c='blue', label='Класс 1')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Задача XOR — классификация шаблонов')
plt.legend()

# График точности
plt.figure()
plt.plot(range(epochs), accuracy)
plt.xlabel('Эпохи')
plt.ylabel('Точность (%)')
plt.title('Точность персептрона')

# Показать диаграмму сетевой архитектуры
print(f"Архитектура персептрона: {perceptron.weights[:-1]} (веса) + {perceptron.weights[-1]} (предвзятость)")

# Show the plots
plt.show()

# Отображение процента правильности
print(f"Процент правильности: {correct_percentage}%")
```

В этой программе класс **Perceptron** остался прежним. Мы добавляем код для создания точек данных для двух классов на основе задачи **XOR**. Точки, принадлежащие к классу **0**, отмечены красным цветом, а точки, принадлежащие к классу **1**, — синим.

После построения двух классов мы строим график точности, аналогичный предыдущей программе. Наконец, мы отображаем архитектуру персептрона, печатая значения весов и смещения.
